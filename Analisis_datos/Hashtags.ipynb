```python
# We import all necessary libraries.
from pathlib import Path
import pandas as pd
import ujson
import unidecode
from collections import Counter
import matplotlib.pyplot as plt
```

GRAFO PRE 8M

```python
# We import the csv file of the graph.
file_pre_8M = r'C:\UOC\TFM\Analisis\pre_8M\analisis_despues_de_limpieza\grafo_pre_8M.csv'
df_pre_8M = pd.read_csv(file_pre_8M)
df_pre_8M.head(5)
```

```python
# We create a dictionary to fill it up with communities' id and all their nodes.
comm_dict_pre_8M = dict()

for index, row in df_pre_8M.iterrows():
    # If the community already exists add just the node id to the community.
    if row[10] in comm_dict_pre_8M:
        comm_dict_pre_8M[row[10]].append(row[0])
    # If the community doesn't exist in the dictionary, create it and add the node id to it.
    else:
        comm_dict_pre_8M[row[10]] = [row[0]]

# We create a function that gets the community id of the user you write as input.
def get_key(val): 
    for key, value in comm_dict_pre_8M.items(): 
         if val in value: 
             return key 
```

```python
# We define where the json files with complete tweets are located.
complete_tweets_pre_8M = r'C:\UOC\TFM\Analisis\pre_8M\analisis_despues_de_limpieza\wordclouds\6_7_M'
```

```python
# We create lists where we are going to save all hashtags used by each community.
hash_comm_14 = []
hash_comm_42 = []
hash_comm_9 = []
hash_comm_1 = []
hash_comm_22 = []
hash_comm_6 = []
hash_comm_8 = []
hash_comm_71 = []
hash_comm_30 = []
hash_comm_49 = []
hash_comm_38 = []


# We iterate over all files located in the defined folder.
for file in Path(complete_tweets_pre_8M).glob('*.json'):
    # For each line of the file.
    for line in open(file, 'r', encoding="utf8"):
        # Skip empty lines.
        if not line.strip (): 
            continue
        # If the line contains a tweet.
        if line.startswith('{"created_at":') or line.startswith('{"contributors":'):
            # Get the name of the tweet originator and its community.
            user_name = ujson.loads(line)['user']['screen_name']
            community = get_key(user_name)
            # Get the hashtags of the tweet.
            tweet_hashtags = ujson.loads(line)['entities']['hashtags']
            # If the community id is one that we want to study, add the text of the hashtags to the correspondent list.    
            if community == 14:
                for hashtag in tweet_hashtags:
                    hash_comm_14.append(hashtag['text']) 
            if community == 42:
                for hashtag in tweet_hashtags:
                    hash_comm_42.append(hashtag['text'])
            if community == 9:
                for hashtag in tweet_hashtags:
                    hash_comm_9.append(hashtag['text'])
            if community == 1:
                for hashtag in tweet_hashtags:
                    hash_comm_1.append(hashtag['text'])
            if community == 22:
                for hashtag in tweet_hashtags:
                    hash_comm_22.append(hashtag['text'])
            if community == 6:
                for hashtag in tweet_hashtags:
                    hash_comm_6.append(hashtag['text'])
            if community == 8:
                for hashtag in tweet_hashtags:
                    hash_comm_8.append(hashtag['text'])
            if community == 71:
                for hashtag in tweet_hashtags:
                    hash_comm_71.append(hashtag['text'])
            if community == 30:
                for hashtag in tweet_hashtags:
                    hash_comm_30.append(hashtag['text'])
            if community == 49:
                for hashtag in tweet_hashtags:
                    hash_comm_49.append(hashtag['text'])
            if community == 38:
                for hashtag in tweet_hashtags:
                    hash_comm_38.append(hashtag['text'])
            
# Print results.
print("Hashtags más utilizados en la comunidad 14: {}\n".format(Counter(hash_comm_14).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 42: {}\n".format(Counter(hash_comm_42).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 9: {}\n".format(Counter(hash_comm_9).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 1: {}\n".format(Counter(hash_comm_1).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 22: {}\n".format(Counter(hash_comm_22).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 6: {}\n".format(Counter(hash_comm_6).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 8: {}\n".format(Counter(hash_comm_8).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 71: {}\n".format(Counter(hash_comm_71).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 30: {}\n".format(Counter(hash_comm_30).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 49: {}\n".format(Counter(hash_comm_49).most_common()[:10]))
print("Hashtags más utilizados en la comunidad 38: {}\n".format(Counter(hash_comm_38).most_common()[:10]))
```
